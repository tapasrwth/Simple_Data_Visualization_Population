```python
#function to show graph population vs year: one country
def population_graph(country_name):
    country_population=list(df[country_name]/10**6)
    plt.plot(df.year,country_population,color='red')
    plt.xticks(df.year[1:])
    plt.xlabel("Year")
    plt.ylabel("Population")
    plt.legend([country_name])
    plt.show()
```

```python
#function to show graph population vs year: all country
def population_graph_all():
    for i in country_list:
        population_graph(i)
```

```python
#function for showing population increase for any country
def population_increase_graph(country_name):
    country_population=list(df[country_name])
    a=[x - country_population[i-1] for i,x in enumerate(country_population)][1:]
    plt.bar(list(df.year[1:]),a)
    plt.xticks(df.year[1:])
    plt.xlabel("Year")
    plt.ylabel("Population_growth")
    plt.legend([country_name])
    plt.show()
```

```python
population_increase_graph("Uganda")
```

```python
#function for showing population increase for all countries
def population_increase_graph_all():
    for i in country_list:
        population_increase_graph(i)
    
```

```python
#compare 2 countries: population and population increase
def population_compare(country_a,country_b):
    plt.figure(figsize=(16,16))
    fig,(ax1, ax2) =plt.subplots(2,1)
    ax1.plot(df.year,df[country_a]/10**6,color='blue')
    ax1.plot(df.year,df[country_b]/10**6,color='red')
    ax1.legend([country_a,country_b])
    ax1.set_ylabel("Population in million")
    ax1.set_xlabel("Year")
    ##-----------------------------###
    country_population1=list(df[country_a]/10**6)
    country_population2=list(df[country_b]/10**6)
    a=[x - country_population1[i-1] for i,x in enumerate(country_population1)][1:]
    b=[x - country_population2[i-1] for i,x in enumerate(country_population2)][1:]
    ax2.bar(list(df.year[1:]),a)
    ax2.bar(list(df.year[1:]),b)
    #plt.xticks(df.year[1:])
    ax2.set_xlabel("Year")
    ax2.set_ylabel("Population_growth in 5 years (in million)")
    ax2.legend([country_a,country_b])
```

```python
population_compare("France","Germany")
```

```python
#regression model: population prediction 
def reg_model(country_name,year_new):
    Year = np.array(df.year).reshape(-1,1)
    reg = LinearRegression()
    reg.fit(Year,df[country_name])
    population_pred = reg.predict([[year_new]])/10**6
    reg_score = reg.score(Year, df[country_name])
    print("model score:",reg_score)
    print("expected population in: {} is {} million".format(year_new,float(population_pred)))
    plt.scatter(Year,df[country_name]/10**6, label='Population')
    plt.plot(Year,(reg.predict(Year))/10**6,color="green",label='Regression_line')
    plt.scatter(year_new,population_pred,marker='*',color="red", label='Predicted value')
    plt.legend(loc='upper left')
    plt.show()
```

```python
#plt.plot(df.year,reg.predict(df[country_name]))
reg_model("Austria",2006)
```

```python
#taking the last row from the df table: population in 2007 and interchanging row and columns
population_2007= (df.iloc[11:,:]).melt()
population_2007_new=population_2007.sort_values(by=["value"],ascending=False).reset_index(drop=True).rename(columns={"variable":"Country","value":"Population"})
#selecting first 10 rows
top_ten=population_2007_new.iloc[:10]
top_ten
```

```python
plt.figure(figsize=(10,6))
sns.barplot(top_ten.Country,top_ten.Population/10**6)
plt.xlabel("Country[top 10]")
plt.ylabel("Population in million in 2007")
plt.show()
```

```python
#rate of change of population compared to population of 1952: percentage of change of each year compared to 1952 where population of 1952 considered as=100
def growth_rate_compared2_1952(country_a,country_b):
    plt.figure(figsize=(10,6))
    growth_rate_a=(df[country_a]/df[country_a].iloc[0]*100)
    growth_rate_b=(df[country_b]/df[country_b].iloc[0]*100)
    sns.lineplot(df.year,growth_rate_a,label=country_a,marker='o')
    sns.lineplot(df.year,growth_rate_b,label=country_b,marker='v')
    plt.xlabel("year")
    plt.ylabel("Population growth rate compared to 1952[=100] in %")
    plt.show
```

```python
growth_rate_compared2_1952("India","Bangladesh")
```

```python
#Percentage change between the current and prior year population 
def population_pct(country_a,country_b):
    plt.figure(figsize=(10,6))
    pct_a=df[country_a].pct_change()
    pct_b=df[country_b].pct_change()
    sns.lineplot(df.year,pct_a,label=country_a,marker='o',color="blue")
    sns.lineplot(df.year,pct_b,label=country_b,marker='v',color="green")
    plt.xlabel("year")
    plt.ylabel("Percentage change in population between the current and previous year")
    plt.show
```

```python
population_pct("India","China")
```

```python
#Growth rate (growth rate=change in population/time) in percent
def population_growth_rate():
    growth_rate_df=pd.DataFrame(columns=["Country","Growth_Rate"])
    for i in country_list:
        growth_rate=((df[i].iloc[11]-df[i].iloc[0])/10**6)/(df.year[11]-df.year[0])
        #GR=format(growth_rate, '.2f')
        growth_rate_df=growth_rate_df.append({"Country":i,"Growth_Rate":growth_rate},ignore_index=True)
     
    growth_rate=growth_rate_df.sort_values(by="Growth_Rate", ascending=False).reset_index(drop=True)
    return growth_rate
```

```python
#countries with highest growth rate
top_growth_rate_countries=population_growth_rate().iloc[:10]
top_growth_rate_countries
```

```python
#growth rate plot for top 10 countries
plt.figure(figsize=(10,6))
sns.barplot(top_growth_rate_countries.Country,top_growth_rate_countries.Growth_Rate)
plt.xlabel("Country")
plt.ylabel("Population growth rate 1952-2007 (dN/dt)")
plt.title("Growth rate:1952-2007")
plt.show
```
